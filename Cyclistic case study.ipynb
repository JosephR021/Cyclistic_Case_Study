```python
#Changing the datatype for both date columns(started_at, started_at)

df.started_at = df.started_at.astype('datetime64[ns]')

df.ended_at = df.ended_at.astype('datetime64[ns]')
```

```python
#checking for false entries in date columns

false_entries = df.loc[(df.started_at > df.ended_at)] 
false_entries.shape
```

```python
false_entries
```

```python
#deleting the fale entries and making a copy of the data

df = df.drop(false_entries.index.to_list()).copy()
```

```python
df
```

```python
#adding a new column to calculate the length of the ride 

df['ride_len_min'] = ((df.ended_at - df.started_at) / pd.Timedelta(minutes=1)).astype('int64')

#creating a new column to extract day

df['day'] = df.started_at.dt.day_name()

#creating a new column to extract month name

df['month_name'] = df.started_at.dt.strftime('%B')

#creating a new column to extract month 

df['month']= df.started_at.dt.month

#creating a new column to extract hour

df['hour'] = df.started_at.dt.hour

#creating a new column to extract day of the week

df['day_of_week'] = df.started_at.dt.dayofweek
```

```python
#Sorting the table by ride_len_min in ascending order

df.sort_values('ride_len_min', ascending=True)
```

```python
#find the percentage of 0 values in ride_len_min

print('The percentage of zero is : ', df.query('ride_len_min == 0').shape[0] / df.shape[0] * 100)
```

```python
#deleting the rows which contains ride_len_min less than 1

#filtering the ride_len_min less than 1
zero_len = df.ride_len_min < 1

#locating the index of ride_len_min less than zero and making a list
zero_len_index = df.loc[zero_len].index.to_list()

#dropping the listm(inplace= False because if it is true it will affect the original dataframe)
df = df.drop(zero_len_index, axis=0 , inplace=False)
```
